```python
pc1 = [56, 48, 40, 32, 24, 16,8,
		  2, 57, 49, 41, 33, 25, 17,
		  9,  1, 58, 50, 42, 34, 26,
		  18,10,  2, 59, 51, 43, 35,
		  62, 54, 46, 38, 30, 22,14,
		  6, 61, 53, 45, 37, 29, 21,
		  13, 5, 60, 52, 44, 36, 28,
		  20, 12,  4, 27, 19, 11, 3]
```

```python
text = '1011010100111010100011010100010100011110100011111101111110010101'
print(text)
print("Length:",len(text))
pc1_text = ''

for i in range(56):
  pc1_text += text[pc1[i]-1]

print(pc1_text)
print("Length after pc1:",len(pc1_text)) 
```

```python
lr = [1, 1, 2, 2, 2, 2, 2, 2, 1, 2, 2, 2, 2, 2, 2, 1]
c = []
d = []

L = pc1_text[:28]
R = pc1_text[28:]

c.append(L[lr[0]:] + L[:lr[0]]) 
d.append(R[lr[0]:] + R[:lr[0]]) 


for i in range(1,16):
  L = c[i-1]
  R = d[i-1]

  c.append(L[lr[i]:] + L[:lr[i]]) 
  d.append(R[lr[i]:] + R[:lr[i]])

keys = []

for i in range(16):
  print("C",i+1,":",c[i])
  print("D",i+1,":",d[i])
  print("")
  keys.append(c[i]+d[i])

for i in range(16):
  print("key",i+1,":",keys[i])
```
